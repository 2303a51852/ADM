```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Load dataset
file_path = "/content/netflix_titles.csv"
data = pd.read_csv(file_path)

# Generate statistical summary of numerical columns
summary = data.describe()
print(summary)
```

```python
# Histogram for 'release_year'
plt.hist(data['release_year'], bins=20, edgecolor='black', color='skyblue')
plt.title('Release Year Distribution')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for 'type' (Movies vs TV Shows)
plt.figure(figsize=(6, 4))
data['type'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Content Type Distribution')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Box plot for 'release_year'
plt.figure(figsize=(6, 4))
plt.boxplot(data['release_year'], patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Release Year')
plt.ylabel('Year')
plt.show()
```

```python
# Count plot for 'rating'
plt.figure(figsize=(8, 4))
sns.countplot(x='rating', data=data, palette='Set2', order=data['rating'].value_counts().index)
plt.title('Content Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
